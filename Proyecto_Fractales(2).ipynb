## Fractales de Madelbrot
### Introducción

Los fractales de Madelbrot, descubiertos por el matemático Benoît Mandelbrot (1924-2010) forman parte de un conjunto en el plano complejo que: 
$$z_{0}=0$$
$$z_{n+1}=z_{n}^2+c$$

	      
   
   
donde __c__ es un número complejo y $z_{0}$ el termino inicial.

Si esta sucesión queda acotada, entonces se dice que c pertenece al conjunto de Mandelbrot, y si no, queda excluido del mismo.

Por otra parte, se sabe que los puntos cuya distancia al origen es superior a 2, es decir,__$\space$ $x^2+y^2\geq4$__ no pertenecen al conjunto.[1] 





```julia
#=Hacemos una funcion que nos devuelva 0 si el punto converge y si no en cuantas iteraciones se supo que no convergia. =#

function mandelbrotConvergence(x,y)
    #Aquí vamos a definir las variables que vamos a ocupar#
    
    N = 50 #Este será el número de iteraciones#
    c = x+im*y # Un número en el campo de los complejos#
    z = 0  #Este es nuestro termino inicial#
    f(x) = z^2+c #Definimos la función 
    iterations  = 0
    convergent = true #=Esta variable la definimos como verdadera al entrar al ciclo for y si la la norma de z es mayos que 2 cambia su valol a falso=#  
    
    for i in 1:N
        z = f(z)
        if(norm(z)>2)
            iterations = i
            convergent = false
            break
        end
    end
    
    iterations = convergent ? 0:iterations # Si comverge nos regresa un cero y si no nos devuelve el número de iteraciones#
    return iterations
end
```

```julia
#Si la sucesión queda acotada pertenecera al conjunto de Madelbrot y si no será excluido#
#=Para c = 1, que no converge, nuestra funcion deberia regresar el numero de iteraciones =#
println("El punto c = 1 converge ? ",mandelbrotConvergence(1,0))

#=Para c = -1 que converge la funcion ha de regresar un cero =#
println("El punto c = -1 converge? ",mandelbrotConvergence(-1,0))
```

```julia
function createMandelbrotMesh(intervalX, intervalY)# Creamos un mallado de números
   mesh = zeros(size(intervalX,1),size(intervalY,1)) # La función size te pide la dimensión, en este caso es 1#
    
   for i in 1:size(intervalX,1)
        for j in 1:size(intervalY,1)
            mesh[j,i] = mandelbrotConvergence(intervalX[i],intervalY[j])/50 #Le aplicamos al mallado nuestra función de convergencia#
        end
    end
   return mesh 
end
```

```julia
using PyPlot 
plt = PyPlot
#Vamos a realizar una función para ver el fractal
function plotMandelbrot(intervalX,intervalY)
    mesh = createMandelbrotMesh(intervalX,intervalY)
    
#Denifimos los limites inferiores y superiores para X y Y para poder usar imshow#    
    minX = intervalX[1]
    maxX = intervalX[size(intervalX,1)]
    minY = intervalY[1]
    maxY = intervalY[size(intervalY,1)]
    
    plt.imshow(mesh,extent = [minX,maxX,minY,maxY])

end
```

```julia

intervalX = linspace(-2,1,500)
intervalY = linspace(-1.5,1.5,500)
title("f(x)=z^2+c")
plotMandelbrot(intervalX,intervalY)
```

### __Método de Newton__

El método de Newton es uno de los más usados para calcular raices de funciones.


$$x_{n}=x_{n-1}-{f(x_{n-1}) \over f'(x_{n-1})}$$

con n en los Naturales es decir n=1,2,3,4......

Esta sucesión converge tan rápido a la raiz como su condición inicial $x_{0}$ este cerca de ella, si nuestra aproximación incial es muy cercana al valor de la raiz, en pocas iteraciones podremos llegar a la raiz real (esto también depende de nuestra tolerancia permitida).

Este método es de los más utilizados debido a que no se restringe unicamente para funciones reales, este método tambien es posible usarlo en funciones complejas.

Un ejemplo de esto es la función:

$$f(z)=z^4-1$$
Entonces para la raiz necesito que:
$$z^4-1=0$$
$$z^4=1$$

Sea ahora $w=z^2$
Entonces:
$$w^2=1$$ 
Así que las soluciones son w=1 y w=-1 ahora resolviendo las z.
$$z^2=1$$  entonces $$z=1 $$  $$  z=-1$$


$$z^2=-1$$ entonces $$z=i $$  $$   z=-i$$


Lo interesante de este método es que para un número aleatorio $z_{0}$ puede tender a cualquiera de las 4 raices y esto tiene una naturaleza fractal. 

Una forma para comprobarlo es generar una función en este caso:
 $$f(z)=z^3 -1$$
 
 Y tomar valores aleatorios en un rango complejo para :
 Sea un número complejo z de tal forma que z=a+ib entonces en este ejemplo usaremos a en [-10,10] y b en [-10,10].
 
 Para esos valores veremos a cual de las tres raices que tiene nuestra función propuesta converge y según sea a la que converge se le asignara un color, así podremos ver la naturaleza fractal de este método en el plano complejo.


```julia
#=En primer lugar haremos la función del metodo de Newton para verificar a que raiz converge cada 
uno de los puntos=#
using SymPy
function newtonMethod(f,g,z)
    MAX_ITERATIONS = 1000
    root = z
    iterations = 0
    convergent = true
    
    while(norm(f(root))>0.001)
       if(iterations<MAX_ITERATIONS) 
       root = root-(f(root)/g(root))
       iterations+=1
            else
            convergent = false
            break
        end
    end
    #=Será importante saber si el método converge y en tal caso a cual de las raices converge.=#

    return[root,convergent]
end
```

```julia
#=Necesitamos saber cuales son las raices de nuestra función, idealmente estaria bien diseñar alguna especie de funcion que las busque, pero como trabajaremos con polinomios, por el momento servirá la función solve de SymPy=#
z = Sym("z")
array = solve(z^3+1 )
```

```julia
#=Esta función busca las raices y les asigna un color aleatorio=#

function assignColors(f::Function)
    g = f(Sym("z"))
    roots = solve(g)
    coloredRoots = []
    for root in roots
        dummyRoot = N(root)
        push!(coloredRoots,[dummyRoot,[rand(),rand(),rand()]])
    end
    return coloredRoots
end
```

```julia
#=Esta funcion se necesita para comparar dos valores cmplejos con un nivel limitado de precisión=#

function areComplexEqual(z1,z2,epsilon)
   if(real(z1-z2)<epsilon && imag(z1-z2)<epsilon)
        return true
    else
        return false
    end
end
```

```julia
areComplexEqual(1.001+1.001*im,1+1*im,0.001)
```

```julia
#=Hacemos una funcion que examine una cuadricula de puntos y nos diga, a que raiz convergen, entonces se le asigna un color especifico dependiendo de la raiz. =#
function createNewtonMesh(f,g,intervalX,intervalY)
    mesh = zeros(size(intervalX,1),size(intervalY,1),3)
    roots = assignColors(f)
    
    for i in 1:size(intervalX,1)
        for j in 1:size(intervalY,1)
            z = intervalX[i]+im*intervalY[j]
            result = newtonMethod(f,g,z)
            #=Si el punto converge por medio del metodo de Newton, entonces asignamos el color 
            que corresponda a la raiz, de lo contrario asignamos el valor [0,0,0] que corresponde en RGB a negro=#
            if(result[2] == true)
                for root in roots
                   if(areComplexEqual(result[1],root[1],0.001))
                       mesh[j, i, :] = root[2]
                    end
                end
            else
                mesh[j, i, :] = [0, 0, 0]
            end
        end
    end
     return mesh
end
```

```julia
using PyPlot
plt = PyPlot

function plotNewton(f,g,intervalX,intervalY)
    mesh = createNewtonMesh(f,g,intervalX,intervalY)
    
#Denifimos los limites inferiores y superiores para X y Y para poder usar imshow#      
    minX = intervalX[1]
    maxX = intervalX[size(intervalX,1)]
    minY = intervalY[1]
    maxY = intervalY[size(intervalY,1)]
    
    plt.imshow(mesh,extent = [minX,maxX,minY,maxY])

end
```
