Ejercicio 2: ¿Como se define una matriz simétrica?



Una matriz simpetrica se define como  $$A=A^{T}$$ , donde  $A^{T}$ es la traspuesta de la matriz $A$ y estas cumplen simetria respecto a su diagonal.

__Ejercicio 3:__ ¿Como puede construir una matriz simétrica a partir de cualquier matriz cuadrada?

Se puede obtener una matriz simétrica a partir de una matriz cuadrada A diagonalizandola tal que 
$$A=PDP^{-1}$$ , donde D es la matriz diagonal con los egeinvectores de A.

$$A=\left(\begin{array}{cc} -1 & 1\\ -1 & -1 \end{array} \right).$$
* Grafique la trayectoria en el espacio de fase.
* Recuerde que $\vec x(t) = e^{A t}\vec x_0$, y necesitamos graficar la trayectoria $\vec x(t)$ en el espacio de fase.

__Nota 1:__ De momento no utilicen `expm`.

__Nota 2:__ El operador `*` funciona para multiplicar matrices.

__Note 3:__ Observen que la matriz $A$ no es simétrica, aqui no se cumple que $w^T=w^{-1}$.

```julia
A = [-1 1;-1 -1]    # Comenzamos definiendo la matriz A
vecs=eigvecs(A)
Diag=diagm(exp.(eigvals(A)))
vals=eigvals(A)
```

```julia
expA(t) = Diag*diagm(exp.(eigvals(A)*t))*inv(Diag)
```

```julia
E(t,x0)=vecs*diagm(exp.(vals*t))*inv(vecs)*x0
```

```julia


using PyPlot
t1 = linspace(0,15,100)             
for i in 1:500                     
    X = []                         
    x0 = [2*rand()-1; 2*rand()-1]   
 
    for k in 1:length(t1)          
        e = E(t1[k], x0)            
        push!(X,e)                  
    end

    listx = map(x->x[1],real(X))   
    listu = map(x->x[2],real(X))
    plot(listx,listu)               
end
title("Espacio fase")




```
