### Bifurcación para sistemas dinamicos


```julia
function bifurcacion(x,r,n) #Comenzamos realizado una función para realizar la bifurcación
    for i in 1:n           #Para iterar
        x = r*x*(1-x)                          
    end
    return x
    end



```

```julia
x_interval=linspace(0,1,50)   #Definimos un intervalo para x
r_interval=linspace(1,4,50) #Definimos un intervalo para r
array1 =[] #En este arreglo se guardaran los valores iterados de x
array2=[] # y en este otro de r
for i in 1:length(x_interval)  #For para llenar los arreglos vacios
    push!(array1,bifurcacion.(x_interval,r_interval[i],1000)) # un push! para ir llenando los arreglos vacios con las iteraciones
    push!(array2,ones(length(x_interval))*r_interval[i])
    end
```

```julia
plot(array2,array1,"red"); #Graficamos nuestro diagrama de bifurcacion 
title("Diagrama de bifurcacion")

legend();
```

```julia

function Lyapunov(x,n,r) #Una funcion para el exponente de Lyaponov
    l=0    
for i in 1:n                                  
    x = r*x*(1-x)                             
    l = l+log(abs(r*(1-2*x)))                 
    end
    return l/n
    end
```

```julia
plot(array2,Lyapunov.(0.5,50,r_interval),"green")
title("Exponente de Lyapunov")
legend();
```
